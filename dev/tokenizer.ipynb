```python
# find out which are not represented by a single token, if not represented by single token add as a special token
gpt2_base = tiktoken.get_encoding('gpt2')
special_tokens = []

for keyword in solidity_keywords:
  enc = gpt2_base.encode(keyword)
  if len(enc) > 1:
    special_tokens.append(keyword)

print(len(special_tokens))
```

```python
simple_solidity = '''// SPDX-License-Identifier: MIT
pragma solidity ^0.8.26;
contract SimpleStorage {
    uint256 public num;

    function set(uint256 _num) public {
        num = _num;
    }
    
    function get() public view returns (uint256) {
        return num;
    }
}'''
```

```python
special_tokens_ids = { token: gpt2_base.n_vocab + i for i, token in enumerate(special_tokens)}
special_tokens_ids
```

```python
enc = tiktoken.Encoding(
  name='sol_gpt2',
  pat_str=gpt2_base._pat_str,
  mergeable_ranks=gpt2_base._mergeable_ranks,
  special_tokens={
        **gpt2_base._special_tokens,
        **special_tokens_ids
    }
)
```

```python
import re

def format_source_code(sc):
  formatted = re.sub(r'/\*[\s\S]*?\*/', '', sc['source_code'])
  formatted = re.sub(r'//[^\n\r]*', '\n', formatted)
  formatted = re.sub(r'[\n\r]{2,}', '\n', formatted)
  formatted = re.sub(r'(\n\s*\n)', '\n', formatted)
  return {'source_code': formatted}
```

```python
def process(dataset_col):
  ids = enc.encode(dataset_col['source_code'], allowed_special=set(special_tokens))
  out = {'ids': ids, 'len': len(ids)}
  return out
```

```python
split_dataset = split_dataset.map(format_source_code, desc='formatting source code', num_proc=5).map(process, remove_columns=['source_code'], desc='tokenizing the source code', num_proc=5)
```
